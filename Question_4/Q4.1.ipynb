```python
with open(r'F:\advent_of_code24\Question_4\Day_4.1.txt','r') as file:
    content = file.read()
```

```python
#140 Lines/Items, 139 index 
xmas_list = content.split('\n')
count_xmas = 0
#XMAS SMAX
for line_index, line in enumerate(xmas_list):
    for element_index, element in enumerate(line):
        #Looking right
        if element == "X" and element_index < len(line) - 3:
           if  line[element_index + 1] == "M":
               if line[element_index + 2] == "A":
                   if line[element_index + 3] == "S":
                       count_xmas += 1
                    #   print (f"right: {line}, {element}")

        #Looking reverse/backwards
        if element == "X" and element_index > 2 :
            if  line[element_index - 1] == "M":
               if line[element_index - 2] == "A":
                   if line[element_index - 3] == "S":
                       count_xmas += 1
                    #   print (f"reverse: {line}, {element}")
                  
    #Looking down
    if line_index < len(xmas_list) - 3:
        for element_index, element in enumerate(line):
            if element == "X":
                next_line = xmas_list[line_index + 1]
                if next_line[element_index] == "M":
                    next_line = xmas_list[line_index + 2]
                    if next_line[element_index] == "A":
                        next_line = xmas_list[line_index + 3]
                        if next_line[element_index] == "S":
                            count_xmas += 1
                            #print (f"down: {line}, {element}")

                if element_index < len(line) - 3:
                    next_line = xmas_list[line_index + 1]
                    if next_line[element_index + 1] == "M":
                        next_line = xmas_list[line_index + 2]
                        if next_line[element_index + 2] == "A":
                            next_line = xmas_list[line_index + 3]
                            if next_line[element_index + 3] == "S":
                                count_xmas += 1
                                #print (f"down/right: {line}, {element}")

                if element_index > 2:
                    next_line = xmas_list[line_index + 1]
                    if next_line[element_index - 1] == "M":
                        next_line = xmas_list[line_index + 2]
                        if next_line[element_index - 2] == "A":
                            next_line = xmas_list[line_index + 3]
                            if next_line[element_index - 3] == "S":
                                count_xmas += 1
                                #print (f"down/left: {line}, {element}")
                            
    #Looking Up
    if line_index > 2:
        for element_index, element in enumerate(line):
            if element == "X":
                prev_line = xmas_list[line_index - 1]
                if prev_line[element_index] == "M":
                    prev_line = xmas_list[line_index - 2]
                    if prev_line[element_index] == "A":
                        prev_line = xmas_list[line_index - 3]
                        if prev_line[element_index] == "S":
                            count_xmas += 1
                            
                if element_index < len(line) - 3:
                    prev_line = xmas_list[line_index - 1]
                    if prev_line[element_index + 1] == "M":
                        prev_line = xmas_list[line_index - 2]
                        if prev_line[element_index + 2] == "A":
                            prev_line = xmas_list[line_index - 3]
                            if prev_line[element_index + 3]  == "S":
                                count_xmas += 1
                                
                if element_index > 2:
                     prev_line = xmas_list[line_index - 1]
                     if prev_line[element_index - 1] == "M":
                        prev_line = xmas_list[line_index - 2]
                        if prev_line[element_index - 2]== "A":
                            prev_line = xmas_list[line_index - 3]
                            if prev_line[element_index - 3]== "S" :
                                count_xmas += 1
                                



print(count_xmas)
```
